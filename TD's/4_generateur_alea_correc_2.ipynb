## 0. Test du $\chi^2$

Dans la cellule de code qui suit, écrire une fonction qui prend en entrée une liste où l'élément d'indice $i$ est le nombre d'occurences de la valeur $i$ obtenu par un procédé inconnu, et une liste contenant la loi à tester (la valeur à l'indice $i$ est la probabilité d'obtenir la valeur $i$), et calcule la valeur du test du $\chi^2$. 


```python
def chi2_test(occurence, proba):
    """
    Calcule la statistique du test du chi2 entre une distribution observée et une distribution théorique.
    occurence: Liste des occurrences observées pour chaque valeur.
    proba: Liste des probabilités théoriques correspondant aux valeurs.
    return: La statistique du test du chi2.
    """
    res = 0
    # nombre total de valeurs
    n = sum(occurence)
    for i in range(len(occurence)):
        o = occurence[i]
        e = n * proba[i]
        res += (o - e)**2 / e
    return res

print(chi2_test([5,5,5,5], [0.25] * 4))
print(chi2_test([4,8,2,9], [0.25] * 4))
```

## 3. Tests sur des séquences

Vous utiliserez la cellule de code ci-dessous pour répondre aux questions.

a. Initialiser une liste avec des nombres aléatoires uniforme sur l'intervalle $[0,1]$ obtenus par la fonction `random`.

b. A partir de cette liste de flottants, générer une liste de bits aléatoires et faites des tests de fréquence pour tester l'uniformité de la distribution.

c. A partir de la liste initiale, générer une liste d'entiers. On va maintenant tester l'uniformité sur cette liste à partir du test des runs croissants: cela consiste à compter le nombre de suites maximales de nombres consécutifs triés en ordre croissant qui sont de taille $k$, pour $k \geq 1$. Par exemple dans la liste suivante:
$$  2,5,6,8,5,1,9,2,3,5,4 $$
la taille des runs croissants est 4, 1, 2, 3, 1. Il y a donc deux runs de taille 1, un run de taille 2, etc.

d. A partir de la liste d'entiers, écrire une fonction qui retourne le nombre de runs croissants de taille $k$ pour $k \geq 1$.

e. Déterminer théoriquement la probabilité que le premier run croissant soit de taille au moins $k$ (vous pouvez commencer par k=1, k=2, k=3, et essayer de généraliser). En déduire la probabilité qu'un run croissant soit de taille $k$.

f. Proposer un test d'uniformité pour la suite des entiers utilisant les runs croissant, et l'implémenter.

g. Tester et commenter le résultat obtenu.




```python
from random import random, seed

def count1(lst):
    # fréquence des 0 et 1
    return [lst.count(0), lst.count(1)]

def count2(lst):
    # compte les fréquence des couples
    res = [0]*4
    # on transforme en string les 0 et 1
    lst_str = list(map(str, lst))
    lst_final = []
    for i in range(len(lst) // 2):
        lst_final.append("".join(lst_str[2*i:2*i+2]))
    for e in lst_final:
        res[int(e, 2)] += 1
    return res

def count3(lst):
    # compte les fréquence des couples
    res = [0]*8
    # on transforme en string les 0 et 1
    lst_str = list(map(str, lst))
    lst_final = []
    for i in range(len(lst) // 3):
        lst_final.append("".join(lst_str[3*i:3*i+3]))
    for e in lst_final:
        res[int(e, 2)] += 1
    return res

def count_run(lst):
    # compte les runs croissants dans la liste selon la taille
    # les runs de taille >= 6 sont comptés dans la même classe
    res = [0]*6
    i = 0
    while i < len(lst):
        taille_run = 1
        while i < len(lst)-1 and lst[i] < lst[i+1]:
            taille_run = min(taille_run + 1, 6)
            i += 1
        res[taille_run - 1] += 1
        i += 2   
    return res
    
    
    
    
seed(123456)
n = 10000
lst_rand = [random() for _ in range(n)]
print(lst_rand)
lst_01 = [0 if e < 0.5 else 1 for e in lst_rand]
print(lst_01)
occurence = count1(lst_01)
print(chi2_test(occurence, [1/2]*2))
occurence = count2(lst_01)
print(chi2_test(occurence, [1/4]*4))
occurence = count3(lst_01)
print(chi2_test(occurence, [1/8]*8))
lst_int = [int(10**10 * e) for e in lst_rand]
print(lst_int)
occurence = count_run(lst_int)
print(chi2_test(occurence, [1/2, 2/6, 3/24, 4/120, 5/720, 1/720]))

```
